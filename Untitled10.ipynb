```python
# Initialize an empty list to store state data
state_data_list = []

# Fetch HTML content from the state URL
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

options = soup.find_all('option')
base_url = "https://results.eci.gov.in/PcResultGenJune2024/"
for option in options:
    state_value = option.get('value')
    state_name = option.text.strip()
    state_url = f"{base_url}partywiseresult-{state_value}.htm"



    # Create a dictionary for each state and append to the list
    state_data = {
        'state': state_name,
        'value': state_value,
        'url': state_url
    }

    # Append data to the list
    state_data_list.append(state_data)

# Now state_data_list contains all the extracted data
print(state_data_list)
for state_data in state_data_list:
    state_value = state_data['value']
    state_url = state_data['url']
    state_name = state_data['state']

    # Fetch HTML content from the state URL
    response = requests.get(state_url)
    soup = BeautifulSoup(response.text, 'html.parser')
        # Initialize the 'parties' key with an empty list
    state_data['parties'] = []
    # Find the table element
    table = soup.find('table')
    if table:
        # Extract table rows
        rows = table.find_all('tr')
        parties_data = []
        for row in rows:
            # Assuming the structure of the table, adjust indexes accordingly
            columns = row.find_all('td')
            if len(columns) >= 4:
                party = columns[0].text.strip()
                won = columns[1].text.strip()
                leading = columns[2].text.strip()
                total = columns[3].text.strip()

                # Append extracted data to the list
                parties_data.append({
                    'state': state_name,
                    'party': party,
                    'won': won,
                    'leading': leading,
                    'total': total,
                    'value': state_value,
                    'url': state_url
                })



        # Store all parties' details in the state data
        state_data['parties'] = parties_data

# Convert the state_data_list to a DataFrame
all_parties_data = []
for state_data in state_data_list:
    all_parties_data.extend(state_data['parties'])

md = pd.DataFrame(all_parties_data)

md
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor

# Function to extract first <tr> inside <tbody> from HTML content using BeautifulSoup
def get_first_tr(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    table = soup.find('table')
    if table:
        tbody = table.find('tbody')
        if tbody:
            first_tr = tbody.find('tr')
            if first_tr:
                # Convert first <tr> to DataFrame with proper headers if needed
                headers = [header.get_text(strip=True) for header in first_tr.find_all('th')]
                if not headers:  # Fallback for tables without <th> headers
                    headers = None
                data = [[data.get_text(strip=True) for data in first_tr.find_all('td')]]
                df = pd.DataFrame(data, columns=headers)
                return df
    return pd.DataFrame()  # Return empty DataFrame if no valid <tr> found

# Function to fetch URL and extract first <tr> inside <tbody>
def fetch_and_process(url):
    page = requests.get(url)
    table_data = get_first_tr(page.content)
    return table_data

# Assuming 'sc' is your DataFrame containing URLs
urls = sc['url'].tolist()

# Concurrently fetch and process URLs
with ThreadPoolExecutor(max_workers=10) as executor:  # Adjust max_workers as needed
    results = list(executor.map(fetch_and_process, urls))

# Concatenate all tables in the results list
df = pd.concat(results, ignore_index=True)

print(df)
```

1. The Average Vote Share per constituency for a winning candidate is 5,76,264 through EVM and 2,348 through postal votes, showing overall in increase vote share across the nation.

```python
# Replace hyphens with 0 and then calculate the mean
df['EVM Votes'] = df['EVM Votes'].replace('-', 0).astype(int)
vote_share=df['EVM Votes'].mean()
# Replace hyphens with 0 and convert the column to numeric type
df['Postal Votes'] = df['Postal Votes'].replace('-', 0).astype(int)

# Now you can calculate the median
median_postal_votes = df['Postal Votes'].median()
print(median_postal_votes)
print(vote_share)
```

2. Average Voting percentage for winning candidate- it shows us that the election remained fair in the end, we infrence this by seeing that at an average for candidates to win in an election, required more than 50% of the votes.

```python
import numpy as np
# Convert '% of Votes' column to numeric type before calculating the range
df['% of Votes'] = df['% of Votes'].str.rstrip('%').replace('-', np.nan).astype(float) # Replace '-' with NaN (Not a Number)

# Now calculate the range
range_percent_votes = df['% of Votes'].max() - df['% of Votes'].min()
print(range_percent_votes)
```
